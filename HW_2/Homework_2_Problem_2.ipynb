Extend the code for mini-batch gradient descent in the notebook linear_regression_gradient_descent.ipynb to the case n=2, that is, the model parameters are the two weights w1 and w2 and the bias term b. To solve this problem, you have to create data points that approximately lie on a 2D plane, display these points, and display the predictions of your model after tuning the parameters with gradient descent.

#Loading the Image Data

```
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm
```

Replaced xs with x1 and x2 for the case of n=2 and altered the function ys to include both x1 and x2 with ys = f(x1, x2)

```
np.random.seed(42)


x1 = 2 * np.random.rand(100, 1)
x2 = 2 * np.random.rand(100, 1)
def f(x1, x2):
  return 4 + 3 * x1 + 3 * x2 + np.random.rand(100, 1)

ys = f(x1, x2)
```

Graph of the points on a plane

```
ax = plt.axes(projection = '3d')
ax.scatter(x1, x2, ys)
plt.xticks(size = 8)
plt.xlabel("$x_1$")
plt.yticks(size = 7)
plt.ylabel("$x_2$")
plt.show()
```

```
# split the data into training and test sets
# train set
train_x1 = x1[:80]
train_x2 = x2[:80]
train_ys = ys[:80]
# test set
test_x1 = x1[80:]
test_x2 = x2[80:]
test_ys = ys[80:]
```

# Linear regression using numpy


```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# initial value for weight w1, w2 and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_x1[i] + w2 * train_x2[i] + b

    grad_w1 = (y_pred - train_ys[i]) * train_x1[i]
    grad_w2 = (y_pred - train_ys[i]) * train_x2[i]
    grad_b  = (y_pred - train_ys[i])

    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b  -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_x1[i] + w2 * test_x2[i] + b - test_ys[i]) ** 2
test_loss /= 20
```

```
test_loss
```

```
pred_ys = w1 * test_x1 + w2 * test_x2 +b
```

#Result
You can see that the predicted values, in blue, are just above or below the actual values we expected, in red.

```
ax = plt.axes(projection = '3d')
ax.scatter(test_x1, test_x2, test_ys, color="blue", label='actual values')
ax.scatter(test_x1, test_x2, pred_ys, color="red", label='prediction values')
plt.xticks(size = 8)
plt.xlabel("$x_1$")
plt.yticks(size = 7)
plt.ylabel("$x_2$")
plt.legend()
plt.show()
```
