#Homework 2 Problem 2
Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

p1 is a minimum of f1
p2 is a maximum of f2
p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.
Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and (b) density plots and (c) contour plots as in density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 in all plots

```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

#Functions
My three functions and requested points for each

$f1(x,y) = x^2+y^2+2x-y$   
>$f1:p1_{minimum}(-1,1/2)=-5/2$

$f2(x,y) = x^2-y^2$
>$f2:p2_{maximum}(0,0)=0$

$f3(x,y) = xy-xy^2$
>$f3:p3_{saddle point}(0,0)=0$

```
def f1(x, y):
  return x**2 + y**2 + 2*y - y

def f2(x, y):
  return -x**2-y**2

def f3(x, y):
  return x*y-x*y**2
```

```
x = np.linspace(-50, 50, 20)
y = np.linspace(-50, 50, 20)
```

#Function 1
$f1(x,y) = x^2+y^2+2x-y$   
>$f1:p1_{minimum}(-1,1/2)=-5/2$


```
X, Y = np.meshgrid(x, y)
Z = f1(x, Y)

ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap='RdGy')
ax.view_init(60, 35)
ax.set_title('f1')

zdata = -5/4
xdata = -1
ydata = 1/2
ax.scatter3D(xdata, ydata, zdata, color='Red')
```

```
plt.contour(X, Y, Z, 50, colors='black')
plt.plot(xdata, ydata, marker='o', markersize=5, color="red")
```

```
plt.imshow(Z, extent=[-50, 50, -50, 50], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
plt.plot(xdata, ydata, marker='o', markersize=5, color="red")
```

#Function 2
$f2(x,y) = x^2-y^2$
>$f2:p2_{maximum}(0,0)=0$

```
X, Y = np.meshgrid(x, y)
Z = f2(x, Y)

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='RdGy')
ax.view_init(60,35)

zdata = 0
xdata = 0
ydata = 0
ax.scatter3D(xdata, ydata, zdata, color='Red')
```

```
plt.contour(X, Y, Z, 50, colors='black')
plt.plot(xdata, ydata, marker='o', markersize=5, color="red")
```

```
plt.imshow(Z, extent=[-50, 50, -50, 50], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
plt.plot(xdata, ydata, marker='o', markersize=5, color="red")
```

#Function 3
$f3(x,y) = xy-xy^2$
>$f3:p3_{saddle point}(0,0)=0$

```
X, Y = np.meshgrid(x, y)
Z = f3(x, Y)

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='RdGy')

zdata = 0
xdata = 0
ydata = 0
ax.scatter3D(xdata, ydata, zdata, color='Red')
```

```
plt.contour(X, Y, Z, 50, colors='black')
plt.plot(xdata, ydata, marker='o', markersize=5, color="red")
```

```
plt.imshow(Z, extent=[-50, 50, -50, 50], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
plt.plot(xdata, ydata, marker='o', markersize=5, color="red")
```
